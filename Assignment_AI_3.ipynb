```
g = { "node 1" : ["node 2","node 3","node 4","node 5"],
          "node 2" : ["node 1", "node 3","node 4","node 5"],
          "node 3" : ["node 1", "node 2","node 4","node 5"],
          "node 4" : ["node 1", "node 2","node 3","node 5"],
          "node 5" : ["node 1", "node 2","node 3","node 4"]
        }
```

```
def generate_edges(graph):
    edges = []
    for node in graph:
        for neighbour in graph[node]:
            edges.append((node, neighbour))

    return edges

edges=generate_edges(g)
```

```
print(len(edges))
```
